```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
def f(y, t, k, m):
    return [y[1], -(k/m)*y[0]]

t = np.linspace(0, 10, 101)
yinit = [0.1, 0]
k = 10.0
m = 1.0
y = odeint(f, yinit , t, args=(k,m))
plt.plot(t,y[:,0],label='Displacement')
plt.plot(t,y[:,1],label='First derivative')
plt.title('Mass and spring system')
plt.xlabel('Time (s)')
plt.ylabel('Displacement (m)')
plt.legend()
plt.grid()
```

```python
def f1(y, t, k, m, b):
    return [y[1], -(k/m)*y[0]-(b/m)*y[1]]

yinit = [0.1, 0]
k = 10.0
m = 1.0
b = 0.5
y_damped = odeint(f1, yinit , t, args=(k,m,b))
plt.plot(t,y_damped[:,0],label='Displacement (Damped)')
plt.plot(t,y_damped[:,1],label='Velocity (Damped)')
plt.title('Mass and Spring system (with Damping)')
plt.xlabel('Time (s)')
plt.ylabel('Displacement (m)')
plt.legend()
plt.grid()
```
